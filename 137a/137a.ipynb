```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0,2*np.pi,100)
ticks = np.linspace(0,2,5)
tick_label = [ f"{tick} $\pi$" for tick in ticks ]
tick_t = ticks * np.pi
```

```python
plt.rcParams['figure.figsize'] = [8, 3]
plt.rcParams['figure.dpi'] = 150

plt.plot(t, np.cos(t/2)**2, label="$\chi_+^z$", c="red")
plt.plot(t, np.sin(t/2)**2, label="$\chi_-^z$", c="blue")

plt.plot(t, (1 - np.sin(t))/2, label="$\chi_+^x$", linestyle="dashed", c="orange")
plt.plot(t, (1 + np.sin(t))/2, label="$\chi_-^x$", linestyle="dashed", c="purple")

plt.axhline(1/2, label="$\chi_\pm^y$", linestyle="dotted", c="green")

plt.xticks(tick_t, tick_label)
plt.grid()

plt.margins(0)
plt.legend()
plt.savefig("blyat.png")
```
