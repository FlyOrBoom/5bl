```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
trial2 = np.genfromtxt('part1-unload-2.csv', delimiter=',')
trial4 = np.genfromtxt('part1-unload-4.csv', delimiter=',')
```

```python
[t2, F2, x2] = trial2[1:].T
[t4, F4, x4] = trial4[1:].T
F2 = -F2
F4 = -F4
```

```python
linest2 = scipy.stats.linregress(x2, F2)
model2 = lambda x: x*linest2.slope + linest2.intercept

linest4 = scipy.stats.linregress(x4, F4)
model4 = lambda x: x*linest4.slope + linest4.intercept

linest2, linest4
```

```python
plt.title("Unloading 2")
plt.xlabel("Displacement (m)")
plt.xlim(0, 0.3)
plt.ylabel("Force (N)")
plt.ylim(0, 4)
plt.scatter(x2, F2, s=0.1, color="black")

domain = np.linspace(0, 0.3, 10)
plt.plot(domain, model2(domain), color="blue", lw=0.5,
         label=f"y = ({linest2.slope:.3f} ± {linest2.stderr:.3f})x + ({linest2.intercept:.3f} ± {linest2.intercept_stderr:.3f})")
plt.grid()
plt.legend()
```

```python
plt.title("Unloading 2")
plt.xlabel("Displacement (m)")
plt.xlim(0, 0.3)
plt.ylim(-0.03, 0.03)
plt.ylabel("Force residuals (N)")
plt.scatter(x2, F2 - model2(x2), s=0.5, color="black")
plt.axhline(0, lw=1, color="blue")
plt.grid()
```

```python
plt.title("Unloading 4")
plt.xlabel("Displacement (m)")
plt.xlim(0, 0.3)
plt.ylabel("Force (N)")
plt.ylim(0, 4)
plt.scatter(x4, F4, s=0.1, color="black")

domain = np.linspace(0, 0.3, 10)
plt.plot(domain, model4(domain), color="blue", lw=0.5,
         label=f"y = ({linest4.slope:.3f} ± {linest4.stderr:.3f})x + ({linest4.intercept:.3f} ± {linest4.intercept_stderr:.3f})")
plt.grid()
plt.legend()
```

```python
plt.title("Unloading 4")
plt.xlabel("Displacement (m)")
plt.xlim(0, 0.3)
plt.ylim(-0.03, 0.03)
plt.ylabel("Force residuals (N)")
plt.scatter(x4, F4 - model4(x4), s=0.5, color="black")
plt.axhline(0, lw=1, color="blue")
plt.grid()
```
